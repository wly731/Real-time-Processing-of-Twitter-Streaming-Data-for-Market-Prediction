```python
def clean_tweet(tweet):
    '''
    input: string
    output: string. Removed whitespaces, hashtags, url
    '''
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())
```

```python
query = '(lang:en "#AMZN" has:hashtags)'
```

```python
# Paginator, improved rate limit issues
# query = '((S&P lang:en) OR (SPY lang:en)) is:verified'
# drom 11/14 to 11/18
i = 0
for tweets in tweepy.Paginator(client.search_all_tweets,
                              query = query,
                              end_time = datetime.datetime(2022, 11, 19),
                              tweet_fields = ['author_id','created_at','public_metrics'],
                              user_fields = ['name','username','id','public_metrics'],
                              start_time = datetime.datetime(2022, 11, 14),
                              max_results = 500):
        
    time.sleep(1)
#     next_token = tweets.json()["meta"]["next_token"]
    if tweets.json()["meta"]["result_count"] != 0:
        tweets_dict = tweets.json()
        tweets_data = tweets_dict['data'] 

        if i==0:
            # Transform to pandas Dataframe
            df = pd.json_normalize(tweets_data)
        else:
            df2 = pd.json_normalize(tweets_data)
            df = pd.concat([df,df2], axis = 0)
#             df = df.append(df2)
            print(df.shape)
        i+=1

df.text = df.text.apply(clean_tweet)
df.to_csv("hashtag/20221114_18_AMZN_hashtag.csv")
```
